```python
# regular expressions: used to match strings of text such as particular characters, words or patterns
# also called regex, regexp
# used to replace, remove, extract specific patterns
# re is best library for regular expressions

# findall: returns a list containing all matches
# search: returns a match object if there is a match anywhere in the string
# split: Returns a match if there is a match anywhere in the string
# sub: replaces one or many matches with a string
```

```python
import re
```

```python
quote = "I saved a life. My own?, Am I a hero? I really can't say, but yes - VVS Varun"
```

```python
re.search("hero",quote)
```

```python
re.split("I",quote)
```

```python
re.sub("I", "you", quote)
```

```python
re.sub("I", "you", quote, count = 1)
```

```python
string = 'I enjoy pranitha in winters'
```

```python
re.findall("[a-m]",string)  #[] is used to search for set of characters
```

```python
string2 = 'I have 122433 dogs'
re.findall("[0-9]", string2)
```

```python
# \ is regex meta character, it searches for something specific, \d gives out digits
re.findall("\d", string2)
```

```python
string3 = 'hello hero helping helpo' # . looks for all characters, expect special charaters
re.findall(".", string3)
```

```python
string3 = 'hello hero helping helpo' # . looks for all characters, expect special charaters
re.findall(".he", string3)
```

```python
string3 = 'hello hero helping helpo' # . looks for all characters, expect special charaters
re.findall("he.o", string3)
```

```python
string3 = 'hello hero helping helpo' # . looks for all characters, expect special charaters
re.findall("he..o", string3)
```

```python
# ^ searches for for very beginning of string and $ searches for very ending of the string

string = 'sunflowwers, sundrop, santa clause, and mushroom are good in spring'
#re.findall("^sun",string)
re.findall("g$",string)
```

```python
# searches for number of occurences, * looks for 0 or more occurence, + looks for atleast 1 occuarnce, ? looks for which has least occuring and print it out

string = 'this thing called a thimble is thick'
re.findall("thi.*s",string)
```

```python
string2 =  'Healping someone is kinding'
re.findall("H.*ing",string2)
```

```python
string2 =  'Hing someone is kind'
re.findall("H.+ing",string2)
```

```python
string = 'hing, heliping someone is kind. I like huing'
re.findall("h.?ing", string)
```

```python
string = 'where are whiine while I look ard'
re.findall("wh.{3}e", string) # same as having ("wh...e")
```

```python
string = 'its a beautiful day in the city with all the beautiful lights.it is lovely' # | will search for either of the 2 words in the string

re.findall("beautiful|lovely", string)
```

```python
# regex Metacharacters oe special cvharacters:

quote = 'My name is VVS Varun. My phone number is 469-961-6124. My email id is vvsv2000@gmail.com'

re.findall("\d", quote)  # similar to ["0-9"]
```

```python
# \w - word character, gives all words from the string, dosent give special characters

re.findall("\w", quote)
```

```python
re.findall("\W", quote) # capital \W is opposite version of \w, gives out all special characters
```

```python
# /A - start of a string

re.findall("\A[A-Z]",quote)
```

```python
# /Z - end of a string

re.findall("[A-Z]/Z", quote)
```

```python
# \b - word boundry

quote = 'I love python is pythonlove a word?'

re.findall(r"\bpython", quote)
```

```python
# \B - Not word boundry
quote = 'I love python is pythonlove a word?'

re.findall(r"\Bython\B", quote)
```
