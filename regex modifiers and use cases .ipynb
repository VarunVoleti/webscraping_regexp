```python
# match object methods:

# group(): return string matched by regex pattern
#groups(): return a tuple containing the strings for all matched subgroups
#start(): return the start position of the match
#end(): return end position of the match
#span(): return a tuple containing the positions of the match

import re
```

```python
string = "It is a BEAUTIFUL day in the city. It is 75 degrees and sunny. "
```

```python
re.search(r"(\b[A-Z]+\b).*(\b\d+\b)",string)
```

```python
result = re.search(r"(\b[A-Z]+\b).*(\b\d+\b)",string)
```

```python
result.groups()
```

```python
result.group(1)
```

```python
result.group(2)
```

```python
result.span()
```

```python
result.start()
```

```python
result.end()
```

```python
# regex modifiers: we can add modifiers to regex to control various aspects of how things are matchd and how it will affect the pattern

# re.I - performs case insensitive matching
#re.L - Interprets words according to current locale. It deals with alphabetic group(\w and \W), word boundry(\b and \B)
#re.M - Makes $ match the end line, at starting^ or ending$
#re.S - make a period(.) match any character, including a newline
# re.U - Interprets letters according to the Unicode character set. It will effect \w\W, \b\B
# re.X - permits 'cuter' regular expression syntax. It ignores whitespaces
```

```python
quote = 'My name is VVS Varun. My phone number is 469-961-6124. My email id is vvsv2000@gmail.com'



```

```python
re.findall(r"\b[a-z]+\b", quote)
```

```python
re.findall(r"\b[a-z]+\b", quote, re.I)
```

```python
target = "Varun like number 3\nPranitha like number 9"
```

```python
re.findall(r"^\w{4}", target)
```

```python
re.findall(r"^\w{5,8}", target, re.M)
```

```python
re.findall(r"\d{1}$", target)
```

```python
re.findall(r"\d{1}$", target, re.M)
```

```python
string = "I like python.\nI like power BI"
```

```python
re.findall(r".+",string)
```

```python
re.findall(r".+",string, re.S)
```

```python
result = re.search(r".+",string)

print(result.group())
```

```python
result = re.search(r".+",string, re.S)
print(result.group())
```

```python
# regex usecases:

random_text = '''
my name is vvs varun. my phone number is 346-414-5122. My email id is vvsv2000@gmail.com
my name is Pranitha. My phone number is 469-961-6124. My email id is pranitha@apple.com
my name is peter. My phone number is 768-098-1768. My email id is peter@wyndham.com
'''
```

```python
re.findall(r"@[a-zA-a]+", random_text)
```

```python
re.findall(r"@([a-zA-a]+)", random_text)
```

```python
re.findall(r"@[\w.\.+]+", random_text)
```

```python
re.findall(r"[\w\.+]+@[\w\.+]+", random_text)
```

```python
re.findall(r"\d{3}-\d{3}-\d{4}",random_text)
```

```python
re.findall(r"\ (\d{3})",random_text)
```

```python
email_list = ['vvsv2000@gmail.com','pranitha@apple.com','peter@wyndham.com']
```

```python

for email in email_list:
    result = re.findall(r"@([a-zA-z]+)",email)
    print(result)
```

```python
domain_list = [re.findall(r"@([a-zA-z]+)",email)for email in email_list]

print(domain_list)
```

```python
import re

# List of Phone Numbers
phone_numbers = ["(123) 456-7890", "(555) 123-4567", "(987) 654-3210", "(800) 555-1212", "(123) 456-7890"]

area_code_pattern = r'(\d{3})'

# List to store extracted area codes:
area_codes = []

# Loop through each phone number and extract area codes:
for phone in phone_numbers:
  match = re.search(area_code_pattern, phone)
  if match:
    area_codes.append(match.group(1))

# Print out the list:
print(area_codes)
```
